```python
from flask import Flask, request, jsonify
import sys, json, datetime, pymysql

app = Flask(__name__)
    
@app.route('/trip', methods=["POST"])
def trip_name():
    
    print(request.get_json())
    loc = request.json.get("action", {}).get("detailParams", {}).get("여행지", {}).get("value", '')
    try:
        conn = pymysql.connect(
            host='0.0.0.0
            user='root',
            port=3306,
            password=*****
            db='db_name
            charset='utf8'
        )

        cur = conn.cursor()
        query = "SELECT pp_sightsee FROM chat_db WHERE pp_pname = %s"
        
        cur.execute(query, (loc))
        data = cur.fetchall()  
        print(data)
        data = data[0][0] # 카카오챗봇에 설정한 파라미터 불러오기.
        
        response = {
            "version": "2.0",
            "template": {
                "outputs": [
                    {
                        "simpleText": {
                            "text": data
                        }
                    }
                ]
            }
        }

    except pymysql.Error as e:
        print(f'에러발생: {e}')
        response = {
            "version": "2.0",
            "template": {
                "outputs": [
                    {
                        "simpleText": {
                            "text": "데이터베이스 연결 오류"
                        }
                    }
                ]
            }
        }

    finally:
        # 데이터베이스 연결 닫기
        if 'conn' in locals() and conn.open:
            conn.close()

    return jsonify(response)




if __name__ == '__main__':
    app.run(host='0.0.0.0', port=80, debug=True)  # web server //groomi ide 사용
```
